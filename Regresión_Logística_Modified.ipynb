```python
# Subir el archivo `Data_Heart.xlsx` a la carpeta raíz de Google Colab:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Configuración para gráficos en Google Colab:
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
sns.set(style="whitegrid")
# Configurar DPI para buena resolución; # Configuración de estilo para seaborn.


# Paso 1:

# Cargar los datos del archivo Excel, asegúrandose de que el archivo esté en la raíz de Colab:
df=pd.read_csv('Data_Heart.csv')



# Mostrar las primeras filas para verificar la carga de datos:
print("Primeras filas del dataset:\n", df.head())

# Información general sobre el dataset:
print("\nInformación del dataset:\n")
df.info()

# Estadísticas descriptivas:
print("\nEstadísticas descriptivas:\n", df.describe())

# Conteo de valores nulos por columnap
print("\nValores nulos por columna:\n", df.isnull().sum())

# Verificar la distribución de la variable objetivo 'Condición':
print("\nDistribución de la variable condition:\n", df['condition'].value_counts())

# Análisis de correlación entre las variables:
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()

# Identificación de valores atípicos mediante diagramas de caja (boxplots) para cada variable numérica:
numeric_columns = ["age", "trestbps", "chol", "thalach", "oldpeak"]
plt.figure(figsize=(15, 10))

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot de {col}")

plt.tight_layout()
plt.show()

# Análisis de relaciones específicas entre atributos de interés usando diagramas de dispersión:
# Ejemplo: relación entre 'chol' y 'trestbps' para explorar tendencias en presión y colesterol:
sns.scatterplot(data=df, x="chol", y="trestbps", hue="condition")
plt.title("Relación entre colesterol y presión arterial en reposo según la condición")
plt.show()

# Relación entre 'Edad' y 'thalach' (frecuencia cardíaca máxima alcanzada):
sns.scatterplot(data=df, x="age", y="thalach", hue="condition")
plt.title("Relación entre Edad y Frecuencia Cardíaca Máxima según la condición")
plt.show()



# Paso 2:

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Tratamiento de valores faltantes:
# Verificar nuevamente valores faltantes por columna:
print("Valores faltantes por columna antes de la limpieza:\n", df.isnull().sum())

# En este caso, si hubiese valores faltantes, podríamos hacer una imputación.
# Vamos a rellenar valores faltantes numéricos con la media de cada columna
df.fillna(df.mean(), inplace=True)

# Verificar que ya no haya valores nulos:
print("\nValores faltantes por columna después de la limpieza:\n", df.isnull().sum())

# 2. Codificación de variables categóricas:
# Variables como 'sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', y 'thal' son categóricas:
# Vamos a convertirlas en variables dummy para la regresión logística:

categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

print("\nDataframe con variables categóricas codificadas:\n", df.head())

# 3. Normalización de variables numéricas:
# Normalizamos los datos para mejorar el entrenamiento de la regresión logística:
numeric_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print("\nDataframe después de la normalización de variables numéricas:\n", df.head())

# 4. Verificación final del dataset preparado:
print("\nInformación final del dataset preparado:\n")
df.info()
print("\nPrimeras filas del dataset final:\n", df.head())



# Paso 3:

from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

# Definir las variables independientes y la variable dependiente:
X = df.drop(columns=['condition'])
y = df['condition']
# "Condición" es la variable objetivo.

# Crear el modelo de regresión logística:
model = LogisticRegression(max_iter=1000, random_state=42)

# Usar RFE para seleccionar características:
# Seleccionaremos las 10 características más relevantes como ejemplo:
selector = RFE(model, n_features_to_select=10)
selector = selector.fit(X, y)

# Ver cuáles son las características seleccionadas:
selected_features = X.columns[selector.support_]
print("Características seleccionadas:\n", selected_features)

# Filtrar el dataset para incluir solo las características seleccionadas:
X_selected = X[selected_features]

print("\nDataset después de la selección de características:\n", X_selected.head())



# Paso 4:

from sklearn.model_selection import train_test_split

# Dividir el dataset con las características seleccionadas en conjuntos de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.3, random_state=42
)

# Verificar las dimensiones de los conjuntos de datos:
print("Dimensiones del conjunto de entrenamiento (X_train):", X_train.shape)
print("Dimensiones del conjunto de prueba (X_test):", X_test.shape)
print("Dimensiones de la variable objetivo de entrenamiento (y_train):", y_train.shape)
print("Dimensiones de la variable objetivo de prueba (y_test):", y_test.shape)



# Paso 5:

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Configurar los hiperparámetros para el modelo de regresión logística:
parametros = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear', 'saga'],
    'max_iter': [100, 200, 500]
}
# Regularización; # Inversa de la regularización; # Algoritmos de optimización;
# Iteraciones máximas para la convergencia


# Configuración del modelo con búsqueda de hiperparámetros
modelo = LogisticRegression()
busqueda = GridSearchCV(modelo, parametros, cv=5, scoring='accuracy')

# Entrenar el modelo con la búsqueda de hiperparámetros:
busqueda.fit(X_train, y_train)

# Seleccionar el mejor modelo y sus hiperparámetros:
modelo_optimizado = busqueda.best_estimator_
print("Mejores hiperparámetros:", busqueda.best_params_)
print("Exactitud en entrenamiento con el mejor modelo:", busqueda.best_score_)



# Paso 6:

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Realizar predicciones con el modelo optimizado en el conjunto de prueba:
y_pred = modelo_optimizado.predict(X_test)

# Calcular métricas de evaluación:
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Mostrar resultados:
print("Exactitud (Accuracy):", accuracy)
print("Precisión (Precision):", precision)
print("Sensibilidad (Recall):", recall)
print("F1-score:", f1)

# Reporte de clasificación completo:
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))

# Generar matriz de confusión:
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusión:
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=["Sin Enfermedad", "Con Enfermedad"], yticklabels=["Sin Enfermedad", "Con Enfermedad"])
plt.xlabel("Predicción")
plt.ylabel("Realidad")
plt.title("Matriz de Confusión del Modelo de Regresión Logística")
plt.show()



# Paso 7:

# Configuración para gráficos en Google Colab:
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
sns.set(style="whitegrid")
# Configurar DPI para buena resolución; # Configuración de estilo para seaborn.


# Matriz de Confusión:
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=["Sin Enfermedad", "Con Enfermedad"], yticklabels=["Sin Enfermedad", "Con Enfermedad"])
plt.xlabel("Predicción")
plt.ylabel("Realidad")
plt.title("Matriz de Confusión del Modelo de Regresión Logística")
plt.show()


from sklearn.metrics import roc_curve, auc
# Predicciones de probabilidad:
y_pred_proba = modelo_optimizado.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Gráfico de la Curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='Curva ROC (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC para el Modelo de Regresión Logística')
plt.legend(loc="lower right")
plt.show()


# Histograma de probabilidades predichas
plt.figure(figsize=(8, 6))
plt.hist(y_pred_proba, bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Probabilidad Predicha de Enfermedad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Probabilidades Predichas")
plt.show()


# Obtener importancia de características
feature_importance = pd.DataFrame({"Características": X_train.columns, "Importancia": abs(modelo_optimizado.coef_[0])})
feature_importance = feature_importance.sort_values(by="Importancia", ascending=False)

# Gráfico de Importancia de Características
plt.figure(figsize=(10, 6))
sns.barplot(x="Importancia", y="Características", data=feature_importance, palette="viridis")
plt.xlabel("Magnitud de Importancia")
plt.ylabel("Características")
plt.title("Importancia de las Características en el Modelo de Regresión Logística")
plt.show()


# Ajustes para errorres y advertencias que no afecta el Proyecto:
# Importar librerías necesarias:
import warnings
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.exceptions import ConvergenceWarning


# Ignorar advertencias de convergencia:
warnings.filterwarnings('ignore', category=ConvergenceWarning)

# Definir el modelo con un número mayor de iteraciones para mejorar la convergencia:
model = LogisticRegression(max_iter=1000)

# Definir el rango de hiperparámetros para la búsqueda en GridSearchCV:
param_grid = {
    'C': [0.1, 1, 10],
    'penalty': ['l2', 'none']
}
# Parámetro de regularización; # Tipo de penalización.

# Configurar GridSearchCV con el parámetro error_score:
grid = GridSearchCV(estimator=model, param_grid=param_grid, error_score=np.nan)

# Ajustar el modelo a los datos:
grid.fit(X_train, y_train)

# Mostrar los mejores parámetros y el mejor puntaje obtenido:
print("Mejores parámetros encontrados:", grid.best_params_)
print("Mejor puntaje obtenido:", grid.best_score_)




```
